```python
# Dataset: https://www.kaggle.com/datasets/nasa/astronaut-yearbook
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('astronauts.csv')
df.head()
```

```python
df.columns
```

```python
df[['mo', 'day', 'year']] = df['Birth Date'].str.split('/', n=2, expand=True)
df.head()
```

```python
df.info()
```

```python

df.astype({'mo': 'str', 'day': 'str'}).dtypes
df.astype({'mo': 'int32', 'day': 'int32'}).dtypes
df["mo"] = df.mo.map("{:02}".format)
df["day"] = df.day.map("{:02}".format)
df.head()
```
